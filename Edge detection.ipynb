```python
import numpy as np             # importing lib.
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
img = cv2.imread("C:\\Users\\saini\\Desktop\\testing\\Edge detection.jpg")
plt.imshow(img)        #reading and showing image file
```

```python
image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)   #converting image to gray
plt.imshow(image,cmap = 'gray')
```

```python
#Defining kernel function also called convolution kernel
kernel = np.array([[-2 ,0,2],             #also called sobel filter
         [-1,0,1],
         [-2,0,2]])
kernel
```

```python
#filtering image
filtered_image = cv2.filter2D(image , -1,kernel)
plt.imshow(filtered_image,cmap = 'gray')
```

```python
#converting filtered image to binary image by creating threshold that set all the pixel of of filtered image
retval ,binary_image = cv2.threshold(filtered_image,100, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(binary_image ,cmap = "gray") #but with high ass filter we get very noisy image
```
